```python
import pandas as pd
```

```python
df = pd.read_csv('./dataset/boston.csv')
```

```python
df
```

```python
data = df.drop('MEDV', axis=1)
target = df['MEDV']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# åˆ›å»ºå­å›¾
nrows = 7  # è¡Œ
ncols = 2  # åˆ—
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 19))
fig.tight_layout(pad=3.0)  # è°ƒæ•´å­å›¾é—´è·

# éåŽ†æ¯ä¸ªç‰¹å¾ï¼Œç»˜åˆ¶æ•£ç‚¹å›¾
for i, column in enumerate(data.columns):
    row = i // ncols  # è®¡ç®—è¡Œç´¢å¼•
    col = i % ncols   # è®¡ç®—åˆ—ç´¢å¼•
    ax = axes[row, col]  # èŽ·å–å½“å‰å­å›¾
    ax.scatter(data[column], target, alpha=0.5, marker='.', s=21, c='red')  # ç»˜åˆ¶æ•£ç‚¹å›¾
    ax.set_title(f'{column} vs MEDV')  # æ ‡é¢˜
    ax.set_xlabel(column)  #  x è½´æ ‡ç­¾
    ax.set_ylabel('MEDV')  #  y è½´æ ‡ç­¾
    ax.grid(True)  # æ·»åŠ ç½‘æ ¼

# éšè—å¤šä½™çš„å­å›¾ï¼ˆå¦‚æžœæœ‰ï¼‰
if len(data.columns) < nrows * ncols:
    for i in range(len(data.columns), nrows * ncols):
        row = i // ncols
        col = i % ncols
        fig.delaxes(axes[row, col])

# æ˜¾ç¤ºå›¾å½¢
plt.show()
```

```python
X = data
y = target
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_dim=13))
```

```python
model.summary()  # x = ax + b
```

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
history = model.fit(X, y, epochs=200)
```
