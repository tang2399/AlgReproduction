```python
import pandas as pd
```

```python
df = pd.read_csv('./dataset/boston.csv')
```

```python
df
```

```python
data = df.drop('MEDV', axis=1)
target = df['MEDV']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 创建子图
nrows = 7  # 行
ncols = 2  # 列
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 19))
fig.tight_layout(pad=3.0)  # 调整子图间距

# 遍历每个特征，绘制散点图
for i, column in enumerate(data.columns):
    row = i // ncols  # 计算行索引
    col = i % ncols   # 计算列索引
    ax = axes[row, col]  # 获取当前子图
    ax.scatter(data[column], target, alpha=0.5, marker='.', s=21, c='red')  # 绘制散点图
    ax.set_title(f'{column} vs MEDV')  # 标题
    ax.set_xlabel(column)  #  x 轴标签
    ax.set_ylabel('MEDV')  #  y 轴标签
    ax.grid(True)  # 添加网格

# 隐藏多余的子图（如果有）
if len(data.columns) < nrows * ncols:
    for i in range(len(data.columns), nrows * ncols):
        row = i // ncols
        col = i % ncols
        fig.delaxes(axes[row, col])

# 显示图形
plt.show()
```

```python
X = data
y = target
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_dim=13))
```

```python
model.summary()  # x = ax + b
```

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
history = model.fit(X, y, epochs=200)
```
